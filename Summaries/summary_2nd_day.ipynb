```python
def check_size(num):
  if num > 20:
    print("Big")
  elif num > 10:
    print("medium")
  else:
    print("Small")
```

```python
check_size(num=15)
```

```python
def return_size(num):
  size = "Small"
  if num > 20:
    size = "Big"
  elif num > 10:
    size = "medium"
  
  return size
```

```python
size = return_size(num=2)
```
